## Task: Predictive Modeling

### Importing libraries 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

### Loading dataset

```python
data = pd.read_csv("Dataset .csv")
data
```

# Modeling

```python
# droping non numeric columns
non_numeric_cols = data.select_dtypes(exclude=[float, int]).columns.tolist()
print("Columns with non-numeric values:")
print(non_numeric_cols)

data.drop(columns=non_numeric_cols, inplace=True)
```

```python
# splitting the data into training and testing sets
X = data.drop(columns = ["Aggregate rating"])
Y = data['Aggregate rating']
x_train, x_test,y_train, y_test = train_test_split(X ,Y ,test_size = 0.20 ,random_state=42)
```

### By  using Linear Regression model

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
# mean square error and r2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Linear Regression - MSE: {mse:.2f}, R2 Score: {r2:.2f}")
```

### By using Decision Tree Regressor model

```python
model = DecisionTreeRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
# mean square error and r2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Decision Tree Regression - MSE: {mse:.2f}, R2 Score: {r2:.2f}")
```

### By using Randon Forest Regressor model

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
# mean square error and r2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Random Forest Regression - MSE: {mse:.2f}, R2 Score: {r2:.2f}")
```

## Task: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Dataset .csv")
data.columns
```

```python
# distribution of ratings using a histogram
plt.figure(figsize=(9, 6))
plt.hist(data['Aggregate rating'], bins=20 , edgecolor='k')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# relationship between price range and rating using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Price range', y='Aggregate rating', data = data , palette='Set3')
plt.xlabel('Price Range')
plt.ylabel('Rating')
plt.title('Relationship between Price Range and Rating')
plt.show()
```

```python
# comparison of average ratings of different cities using a bar plot
plt.figure(figsize=(12, 6))
average_ratings_by_city = data.groupby('City')['Aggregate rating'].mean()
sns.barplot(x=average_ratings_by_city.index, y=average_ratings_by_city.values, palette='Paired')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Ratings by City')
plt.xticks(rotation=90)
plt.show()
```

```python
# comparison of average ratings of different cuisines using a bar plot
plt.figure(figsize=(12, 6))
average_ratings_by_cuisine = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
sns.barplot(x=average_ratings_by_cuisine.index, y=average_ratings_by_cuisine.values, palette='Blues')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.title('Average Ratings by Cuisine')
plt.xticks(rotation=90)
plt.show()
```

```python
# visualization of the relationship between various features and the target variable
plt.figure(figsize=(12, 6))
sns.boxplot(x='Cuisines', y='Aggregate rating', data = data , palette='Set2')
plt.xlabel('Cuisine')
plt.ylabel('Rating')
plt.title('Relationship between Cuisine and Rating')
plt.xticks(rotation=90)
plt.show()
```
